```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Text

```python
image = cv2.imread('simpsons_frame0.png')
```

```python
plt.imshow(image)
```

```python
img2gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img2gray,cmap='gray')
```

```python
ret3,th3 = cv2.threshold(img2gray,10,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(th3,cmap='gray')
```

```python
figure_size = 5
```

```python
new_image = cv2.blur(th3,(figure_size, figure_size))
```

```python
plt.imshow(new_image,cmap='gray')
```

```python
ret3,th4 = cv2.threshold(new_image,10,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(th4,cmap='gray')
```

```python
figure_size = 5
```

```python
mask = cv2.blur(th4,(figure_size, figure_size))
```

```python
plt.imshow(mask,cmap='gray')
```

```python
image_final = cv2.bitwise_and(image, image, mask=mask)
```

```python
plt.imshow(image_final,cmap='gray')
```

# Shirt

```python
image = cv2.imread('shirt_sample.jpg')
plt.imshow(image)
```

```python
image = cv2.bitwise_not(image)
fig, ax = plt.subplots(figsize=(20,10))
plt.imshow(image, cmap='gray')
```

```python
img2gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
fig, ax = plt.subplots(figsize=(20,10))
plt.imshow(img2gray, cmap='gray')
```

```python
img2gray = cv2.Laplacian(img2gray, cv2.CV_16S, ksize=3)
fig, ax = plt.subplots(figsize=(20,10))
plt.imshow(img2gray, cmap='gray')
```

```python
img2gray = cv2.blur(img2gray, (3, 3))
fig, ax = plt.subplots(figsize=(20,10))
plt.imshow(img2gray, cmap='gray')
```

```python
new_image1 = cv2.Laplacian(img2gray, cv2.CV_16S, ksize=5)
fig, ax = plt.subplots(figsize=(20,10))
plt.imshow(new_image1, cmap='gray')
```

```python
ret3,th = cv2.threshold(new_image1,255,255,cv2.THRESH_BINARY)
fig, ax = plt.subplots(figsize=(20,10))
plt.imshow(th, cmap='gray')
```

```python
new_image2 = cv2.blur(th, (3, 3))
fig, ax = plt.subplots(figsize=(20,10))
plt.imshow(new_image2, cmap='gray')
```

```python
ret3,th = cv2.threshold(new_image2,150,255,cv2.THRESH_BINARY)
fig, ax = plt.subplots(figsize=(20,10))
plt.imshow(th, cmap='gray')
```
